```python
import pandas as pd # selecting of data
data = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107, 108],
    "Name": ["Amit", "Priya", "Rahul", "Sneha", "Arjun", "Meera", "Karan", "Pooja"],
    "Age": [17, 18, 17, 19, 18, 17, 19, 18],
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai", "Kolkata", "Mumbai", "Chennai", "Delhi"],
    "Marks": [85, 92, 76, 88, 95, 67, 73, 90] }
df = pd.DataFrame(data)
print("complete Dataset:")
print(df)
```

```python
print(df["Name"])
print(df[["Name","Marks"]])#select multiple columns
```

```python
print(df.iloc[0]) # 1 row
```

```python
print(df.iloc[:3])
```

```python
#specific row and column(row 2,column 3)
print(df.iloc[2,3])
```

```python
print(df.loc[4]) # iloc is by default integer
```

```python
#filtering of data
print(df[df["Marks"] > 85]) # df 2 times 1 for whole data 2 nd for
```

```python
print(df[(df["Marks"]>80) & (df["Age"] < 19)])
```

```python
print(df[(df["City"]== "Delhi") | (df["City"] == "Mumbai")])
```
