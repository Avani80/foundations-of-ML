```python
import numpy as np
# columns : [age,math  marks , science marks]
data = np.array([
 [18,85,78],
 [19,92,88],
 [17,76,95],
 [18,65,70],
 [20,90,85]
])
data.shape #get the shape of the matrix
```

```python
# find the average age of the students
np.mean(data[:,0])
```

```python
#extract math marks of all students
data[:,1]
```

```python
#find teh highest science max
np.max(data[:,2])
```

```python
# get details of the student who scored more then 90 maths
data[data[:,1] > 90]
```

```python
#increase maths marks of all students by 5
data[:,1] = data[:,1]+5
data
```

```python
#calculate the average of marks in each subject column wise
data.mean(axis=0)
```

```python
#replace all science marks < 75  with 0
data[data[:,2] < 75,2] = 0
data
```

```python
# get data of students who scored at least 80 in both subjects
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```
